```python
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <title>Quiz & Joke Fun</title>
  <style>
    body {
      margin: 0;
      font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
      background: linear-gradient(135deg, #ff9a9e, #fad0c4, #fad0c4, #a18cd1, #fbc2eb);
      background-size: 300% 300%;
      animation: gradient 10s ease infinite;
      color: #333;
      text-align: center;
    }

    @keyframes gradient {
      0% { background-position: 0% 50%; }
      50% { background-position: 100% 50%; }
      100% { background-position: 0% 50%; }
    }

    .container {
      padding: 50px;
    }

    .card {
      background: #ffffffcc;
      border-radius: 10px;
      padding: 20px;
      margin: 20px auto;
      max-width: 500px;
      box-shadow: 0 4px 20px rgba(0,0,0,0.1);
    }

    button {
      background-color: #ff7eb3;
      border: none;
      padding: 10px 20px;
      margin-top: 10px;
      color: white;
      font-weight: bold;
      cursor: pointer;
      border-radius: 5px;
    }

    button:hover {
      background-color: #ff65a3;
    }
  </style>
</head>
<body>
  <div class="container">
    <h1>ðŸŽ‰ Quiz & Joke Time! ðŸ˜‚</h1>

    <div class="card">
      <h2>ðŸ§  Quiz</h2>
      <p id="quiz-question">Loading quiz...</p>
      <button onclick="getQuiz()">New Quiz</button>
    </div>

    <div class="card">
      <h2>ðŸ¤£ Joke</h2>
      <p id="joke-setup">Loading joke...</p>
      <p id="joke-punchline"></p>
      <button onclick="getJoke()">New Joke</button>
    </div>
  </div>

  <script>
    function getQuiz() {
      fetch('https://opentdb.com/api.php?amount=1&type=multiple')
        .then(res => res.json())
        .then(data => {
          const question = data.results[0].question;
          document.getElementById('quiz-question').innerHTML = decodeHTMLEntities(question);
        });
    }

    function getJoke() {
      fetch('https://official-joke-api.appspot.com/random_joke')
        .then(res => res.json())
        .then(data => {
          document.getElementById('joke-setup').textContent = data.setup;
          document.getElementById('joke-punchline').textContent = data.punchline;
        });
    }

    function decodeHTMLEntities(text) {
      var txt = document.createElement("textarea");
      txt.innerHTML = text;
      return txt.value;
    }

    // Load initial content
    getQuiz();
    getJoke();
  </script>
</body>
</html>
```
